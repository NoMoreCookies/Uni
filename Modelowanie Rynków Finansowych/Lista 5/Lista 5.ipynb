
Zadanie 1

 Na rynku występują jedynie akcje dwóch spółek Alfa i Omega.
 Cena jednej akcji spółki Alfa wynosi 30 PLN. 
 Liczba akcji spółki Alfa równa jest 10 000. 
 Współczynnik beta akcji spółki Alfa wynosi 1,1. 
 Cena jednej akcji spółki Omega wynosi 10 PLN. 
 Liczba akcji spółki Omega jest równa 20 000. 
 Wyznacz wartość współczynnika beta dla akcji spółki Omega.



```python

nA = 10000

cA = 30

bA = 1.1

cB = 10

nB = 20000

wR = nA * cA + nB * cB

wA = nA * cA

wB = nB * cB

# Wyprowadzenie tego w notatkach

# 1 = wA/wR * bA + wB/wR * bB

bB = (1 - wA/wR * bA) * wR/wB

print(bB)


```

Zadanie 8

Cena spot srebra wynosi 48 USD za uncję. Koszty przechowywania wynoszą 0.64 USD
za uncję na rok, płatne kwartalnie z góry. Zakładając, że stopa procentowa w USD
wynosi 4.5% na rok (kapitalizacja ciągła) dla wszystkich okresów, oblicz cenę srebra
w kontraktach futures z dostawą za 9 miesięcy.

```python
import math


S0 = 48.0          
r = 0.045          
T = 0.75           
storage_cost = 0.64
payments_per_year = 4

Z = storage_cost / payments_per_year 

s_times = [0.0, 0.25, 0.50]

discounted_storage = sum(Z * math.exp(-r * s) for s in s_times)

F0T = (S0 + discounted_storage) * math.exp(r * T)

print("Cena :", F0T)
```
